```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download ('words')
nltk.download('treebank')
from nltk import word_tokenize,pos_tag
```

```python
text1 = " The Ottawa Senators came back from a 10-day break with an impressive 5-3 win over the Toronto Maple Leafs on Saturday night. Led by Shane Pinto, who had a goal and two assists, the Senators (21-25-2) also got goals from Claude Giroux, Vladimir Tarasenko, Josh Norris and Ridly Greig. Joonas Korpisalo made 32 saves. Auston Matthews, Matthew Knies and Max Domi scored for the Leafs (26-15-8). Martin Jones stopped 28 shots. Trailing 4-2 to start the third period, the Leafs made it a one-goal game at 5:12 thanks to a defensive breakdown by Ottawa leaving Domi alone out front for his fifth of the season. Greig added an empty-net goal with 5.1 seconds remaining. Morgan Reilly took exception to the slapshot Greig used for the empty-net goal and cross-checked Greig up high."
```

```python
print(text1)
```

```python
tokens1 = word_tokenize(text1)
tag1=pos_tag(tokens1)
print(tag1)

ne_tree1 = nltk.ne_chunk(tag1)
print(ne_tree1)
```

PARTY: Toronto Maple Leafs, Ottawa Senators
PERSON: Auston Matthews, Matthew Knies, Max Domi, Martin Jones, Morgan Reilly, Josh Norris, Ridly Greig, Shane Pinto, Vladimir Tarasenko, and Joonas Korpisalo
Identified Coins:


Every word or token in the text has a matching part-of-speech tag.
For instance, ('The', 'DT') denotes that 'The' is classified as a determiner (DT), ('Ottawa', 'NNP') denotes that 'Ottawa' is a proper noun (NNP), and this pattern continues.

Within the text, named entities are identified and indicated along with the appropriate entity kinds.
For instance, '(ORGANIZATION Ottawa Senators)' denotes the recognition of 'Ottawa Senators' as an organization, and '(PERSON Shane Pinto)' denotes the recognition of 'Shane Pinto' as an individual.
